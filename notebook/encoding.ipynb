```python
data=pd.read_csv("https://raw.githubusercontent.com/mohitsandhuria/Data_Science-Machine_learning/main/data_set/adult.csv",keep_default_na=False,na_values='?',names=['age',"workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","salary"])
```

```python
data
```

```python
labler={}
cat_cols=[var for var in data.columns if data[var].dtype=='O']
for i in cat_cols:
    le=LabelEncoder()
    le=le.fit(data[i])
    data[i]=le.transform(data[i])
    labler[i]=le
```

```python
labler
```

```python
for i in cat_cols:
    data[i]=labler[i].inverse_transform(data[i])
```
